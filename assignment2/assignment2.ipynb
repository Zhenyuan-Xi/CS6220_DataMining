```python
pd.set_option("display.max_columns", 50) 
url2 = "https://ww2.amstat.org/publications/jse/v19n3/Decock/AmesHousing.txt" 
df2 = pd.read_csv(url2, sep="\t")
```

```python
df2.head()

# 79 features, 36 numeric
```

```python
df2_numeric = df2.drop(df2.columns[[0, 1, 81]], axis=1)._get_numeric_data()
df2_numeric.describe()
```

```python
fig, axes = plt.subplots(12, 3, figsize=(20,38))
for i in range(0, 12):
    for j in range(0, 3):
        df2_numeric.iloc[:, [3*i+j]].hist(grid=False, ax=axes[i,j])
        fig.subplots_adjust(hspace=0.5)
```

```python
fig2, axes2 = plt.subplots(12, 3, figsize=(20,100))
for i in range(0, 12):
    for j in range(0, 3):
        df2_numeric.iloc[:, [3*i+j]].boxplot(grid=False, ax=axes2[i,j])
        fig2.subplots_adjust(hspace=0.5)
```
